```python
import os
import pandas as pd
from sqlalchemy import create_engine

import pymysql
pymysql.install_as_MySQLdb()
```

```python
data_read = os.path.join('Resources', 'clean_hawaii_measurements.csv')
data_read1 = os.path.join('Resources', 'clean_hawaii_stations.csv')
data_measurements = pd.read_csv(data_read)
data_measurements.head()
                          
```

```python
data_stations = pd.read_csv(data_read1)
data_stations.head()
```

```python
from sqlalchemy import create_engine, ForeignKey
from sqlalchemy import Column, Date, Integer, String
from sqlalchemy.ext.declarative import declarative_base

Base = declarative_base()

class Measurements(Base):

    __tablename__ = "measurement"
    id = Column(Integer, primary_key=True)
    station = Column(String(225))
    date = Column(Integer)
    prcp = Column(Integer)
    tobs = Column(Integer)
    
class Station(Base):

    __tablename__ = "station"
    id = Column(Integer, primary_key=True)
    station = Column(String(225))
    name = Column(String)
    latitude = Column(Integer)
    longitude = Column(Integer)
    elevation = Column(Integer)


engine = create_engine('sqlite:///Resources/hawaii.sqlite', echo=True)

Base.metadata.create_all(engine)



```

```python
measure_data = data_measurements.to_sql('measurement', engine, if_exists='replace', index ='True')
measure_data
```

```python
station_data = data_stations.to_sql('station', engine, if_exists='replace', index ='True')
station_data
```
